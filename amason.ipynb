```python

average_price = df['price'].mean()
df['price'] = df['price'].fillna(average_price)
df['ratings'] = df['ratings'].fillna(df['ratings'].mean())
df['ranks']=df['ranks'].fillna(df['ranks'].mean())
df['no_of_reviews']=df['no_of_reviews'].fillna(df['no_of_reviews'].mean())
```

```python
df['author'] = df['author'].fillna('Unknown')
df['genre'] = df['genre'].fillna('Unknown')
df['cover_type']=df['cover_type'].fillna("unknown")
df['title']=df['title'].fillna("unknown")
```

```python
df.isnull().sum()
```

```python
df.boxplot(column='price')
plt.show()
```

```python
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR
# Step 3: Clip(keeps all your data but caps extreme values) the values
df['price'] = df['price'].clip(lower=lower_fence, upper=upper_fence)
```

```python
df.boxplot(column='price')
plt.show()
```

```python
df.boxplot(column= 'no_of_reviews')
plt.show()
```

```python
for col in ['no_of_reviews']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 1.5 * IQR
    upper_fence = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower_fence, upper_fence)
```

```python
df.boxplot(column= 'no_of_reviews')
plt.show()
```
